# RMSE

```python
def rmse(y, y_pred):
    error = y-y_pred
    se = error**2
    mse = se.mean()
    return mse**0.5


rmse(y_train, y_pred)
```

# categorical variables

Number of doors

```python
df_train.dtypes.sort_values()
```

```python
categories = ["engine_fuel_type","make","transmission_type","driven_wheels","market_category","vehicle_size","vehicle_style"]

categories_dict = {}

for cat in categories:
    counts = df[cat].value_counts().head().index.tolist()
    categories_dict[cat] = counts

categories_dict
```

```python
def prepare_X(df):
    df = df.copy()


    df["age"] = df.loc[:, "year"].max() - df.loc[:, "year"]
    doors_cols = []
    for i in [2, 3, 4]:
        df.loc[:,f"number_of_doors_{i}"] = (df.loc[:, "number_of_doors"] == i).astype(int)
        doors_cols.append(f"number_of_doors_{i}")

    for k,v in categories_dict.items():
        for cat in v:
            df.loc[:, f"{k}_{cat}"] = (df.loc[:, k] == cat).astype(int)
            doors_cols.append(f"{k}_{cat}")

    base_cols = ["engine_hp", "engine_cylinders", "highway_mpg", "city_mpg", "popularity","age"] + doors_cols
    
    df.loc[:, base_cols] = df.loc[:, base_cols].fillna(0)
    return df[base_cols].values



df_train
```

```python
X_train = prepare_X(df_train)

w0, w = train_linear_regression(X_train, y_train)
y_pred = w0 + X_train.dot(w)

rmse(y_train, y_pred)
```

# Regularization

```python
def train_linear_regression(X, y, r=0.001):

    ones = np.ones(X.shape[0])
    X1 = np.column_stack([ones, X]) # We add the bias term



    XTX = X1.T.dot(X1)
    XTX = XTX + r * np.eye(XTX.shape[0]) # regularization
    XTX_inv = np.linalg.inv(XTX)

    w_full = XTX_inv.dot(X1.T).dot(y)
    return w_full[0], w_full[1:]


train_linear_regression(X, y)
```

```python
def modeling_pipeline(df_train, y_train, r):    
    X_train = prepare_X(df_train)

    w0, w = train_linear_regression(X_train, y_train, r)
    y_pred = w0 + X_train.dot(w)

    print("RMSE:", rmse(y_train, y_pred))
    print("w0:", w0)
    print("w:", w)
```

```python
for i in [0, 0.0001, 0.001, 0.01, 0.1, 1, 10]:
    print("r =", i)
    modeling_pipeline(df_train, y_train, i)
    print("-"*100)
```
